## Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Flipkart MObile.csv')  # load data set
print(data)
```

```python
data.drop(['model'], axis = 1, inplace = True)
data.drop(['brand'], axis = 1, inplace = True)
data.drop(['base_color','processor','screen_size'], axis = 1, inplace = True)
```

```python
X = data.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = data.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
lr = LinearRegression()  # create object for the class
lr.fit(X, Y)    # perform linear regression
Y_pred = lr.predict(X)  # make predictions
```

```python
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='yellow')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from factor_analyzer import FactorAnalyzer
```

```python
data = pd.read_csv('Flipkart Mobile.csv')
print(data)
```

```python
data.columns
```

```python
data.info()
```

```python
# Remove the unnamed column
data = data.iloc[0:,1:16]
```

```python
data.head()
```

```python
data.drop(['model'], axis=1, inplace=True)
data.drop(['brand'], axis=1, inplace=True)
data.drop(['base_color','processor','screen_size'], axis = 1, inplace = True)
```

```python
# Create a factor analyzer variable and perform factor analysis
fa = FactorAnalyzer(n_factors=6, rotation="varimax")
```

```python
data = data.dropna()
```

```python
fa.fit(data)
```

```python
# Get the loadings
loadings = fa.loadings_
```

```python
# Get the Eigenvector and Eigenvalues
ev, v = fa.get_eigenvalues()
```

```python
# Do a scree plot
xvals = range(1, data.shape[1]+1)
```

```python
plt.scatter(xvals, ev)
plt.plot(xvals, ev)
plt.title('Scree Plot')
plt.xlabel('Factor')
plt.ylabel('Eigenvalue')
plt.grid()
plt.show()
```
